```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
def get_year_quarter(month_num):
    if 12 <= month_num and month_num <= 2:
        return 1
    if 3 <= month_num and month_num <= 5:
        return 2
    if 6 <= month_num and month_num <= 8:
        return 3
    if 9 <= month_num and month_num <= 11:
        return 4
    return None
```

```python
df['quarter'] = df['Date'].dt.quarter
```

```python
cols_to_exclude = ["Date", "Rooms", "Bedroom", "Bathroom", "Car", "quarter"]
max_unique_count = 150
for col in df.columns:
    if df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        df[col] = df[col].astype('category')
```
